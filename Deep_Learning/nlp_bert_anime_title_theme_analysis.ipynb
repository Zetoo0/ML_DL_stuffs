```python
##descript
from sklearn.model_selection import train_test_split
import datasets
from datasets import Dataset, DatasetDict
from transformers import BertTokenizer
desc_df.rename(columns = {'target':'label'}, inplace = True)

display(desc_df.text.to_string())

train_df, val_df = train_test_split(desc_df, random_state = 42, train_size = 0.8, stratify = desc_df['label'])


tds = Dataset.from_pandas(train_df)
vds = Dataset.from_pandas(val_df)


ds = DatasetDict()

ds['train'] = tds
ds['validation'] = vds

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
```

```python
class LabelEncoder:
    def __init__(self):
        self.labels_to_int = {}

    def encode(self, labels):
        prev_label = 0
        encoded_labels = []

        for label in labels:
            if label not in self.labels_to_int:
                self.labels_to_int[label] = prev_label
                encoded = prev_label
                prev_label += 1
            else:
                encoded = self.labels_to_int[label]

            encoded_labels.append(encoded)

        return encoded_labels


label_encoder = LabelEncoder()
```

```python
from transformers import DataCollatorWithPadding
import evaluate
import numpy as np
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
from datasets import ClassLabel

#labels = ClassLabel(names_file='labels.txt')
def preprocess_function(examples):
    tokens = tokenizer(examples["text"],truncation=True,padding=True)
    tokens["label"] = label_encoder.encode(examples["label"])
    return tokens


tokenized_ds = ds.map(preprocess_function, batched=True)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)


accuracy = evaluate.load("accuracy")


def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)


id2label = {0: "Mecha", 1: "Isekai", 2: "Anthropomorphic",3: "Martial_Arts",4:"Mythology",5:"Psychological",6:"Music",7:"School",8:"Super_Power",9:"Gag",10:"Adult",11:"Parody",12:"Military",13:"Historical",14:"Idol(Female)",15:"Idol(Male)",16:"Space"}
label2id = {"Mecha": 0, "Isekai": 1, "Anthropomorphic": 2, "Martial_Arts" : 3, "Mythology":4, "Psychological":5, "Music":6,"School":7,"Super_power":8,"Gag":9,"Adult":10,"Parody":11,"Military":12,"Historical":13,"Idol(Female)":14,"Idol(Male)":15,"Space":16}


model = AutoModelForSequenceClassification.from_pretrained(
    "google-bert/bert-base-uncased", num_labels=17, id2label=id2label, label2id=label2id
)
model
```

```python
##descrrir

training_args = TrainingArguments(
    output_dir="my_awesome_model",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    push_to_hub=False,
)

print(tokenized_ds["train"])

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_ds["train"],
    eval_dataset=tokenized_ds["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

trainer.model
#trainer.train()
```
